```python
import requests
from datetime import datetime

requisicao = requests.get("https://economia.awesomeapi.com.br/last/USD-BRL,EUR-BRL,BTC-BRL")

requisicao_dic = requisicao.json()
cotacao_dolar = requisicao_dic["USDBRL"]["bid"]
cotacao_euro = requisicao_dic["EURBRL"]["bid"]
cotacao_btc = requisicao_dic["BTCBRL"]["bid"]



print(f"Cotação Atualizada. {datetime.now()}\nDólar: R${cotacao_dolar}\nEuro: R${cotacao_euro}\nBTC: R${cotacao_btc}")
```

```python
#cotação Bitcon no mês de agosto de 2023
# Crie um intervalo de datas de 01/01/1960 a 31/12/2023
data_inicial = '2023-07-01'
data_final = '2023-07-31'
dados_diarios = pd.date_range(start=data_inicial, end=data_final, freq='D')
```

```python
dados_diarios_df = pd.DataFrame({'data': dados_diarios})
display(dados_diarios_df)
```

```python
dados_diarios_df['bitcoin(R$)'] =1
for i,valor_bit in enumerate(lista_cotacoes_btc):
    dados_diarios_df.loc[i,'bitcoin(R$)']= valor_bit

    
display(dados_diarios_df)
```

```python
import matplotlib.dates as mdates

# Set the figure size and style (selecionar o estilo e o tamanho)
plt.figure(figsize=(18, 7))
sns.set(style="whitegrid")

x=dados_diarios_df['data']
y=dados_diarios_df['bitcoin(R$)']


ax = sns.lineplot(x=x, y=y, color='blue', marker='o', markersize=8, label='Line Plot')

# Format the x-axis labels (formato da data no gráfico)
date_format = mdates.DateFormatter("%d/%m/%Y")  # Set the desired date format
ax.xaxis.set_major_formatter(date_format)

# Rotate the x-axis labels for better readability (rotacionar o eixo x)
plt.xticks(rotation=70)

# Annotate the data values within the plot (plotar valores dentro do gráfico)
for x_value, y_value in zip(x, y):
    ax.annotate(f"R$ {y_value:.2f}",  # Format the value as currency with two decimal places
                xy=(x_value, y_value),
                xytext=(0, 15),
                textcoords='offset points',
                ha='center',
                rotation=90,
                fontsize=15.5,
                fontweight='bold')  # Make the text bold

# Set labels and adjust other settings
plt.xlabel('Mês', fontsize=20)
plt.ylabel('bitcoin(R$)', fontsize=20)
plt.ylim(top=160000)
plt.xticks(fontsize=17)
plt.yticks(fontsize=17)
plt.margins(x=0.03)

# Set a title
plt.title('Cotação da Bitcoin de Agosto de 2023. Fonte: Awesomeapi', fontsize=30, color='red', fontweight='bold')

# Show the plot
plt.show()
```
